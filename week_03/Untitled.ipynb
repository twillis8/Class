```python
class a_sample_class:

    #default/basic constructor
    def __init__(self):
        self.var = 0

    #an instance method for printing
    def print_vars(self):
        print(self.var)

#create instance of class
obj = a_sample_class()

#call instance method to get value for var
obj.print_vars()
```

```python
class a_sample_class:

    #parameterized constructor
    def __init__(self, one, two):
        self.first = one

    #an instance method for printing
    def print_vars(self):
        print(self.first)

#create instance of class
obj = a_sample_class()
obj.print_vars(2,8)
```

```python
class a_sample_class:
    class_attr = 5

    def __init__(self, one, two):
        #instance attribute
        self.first = one
        self.second = two

#object is instance of class
obj = a_sample_class(17,42)

#instance attribute as property of object
print(obj.first)

#class attribute as property of object
print(obj.class_attr)

#instance attribute as property of class
print(a_sample_class.first)

#class attribute as property of class
print(a_sample_class.class_attr)
```

# Instance Attributes and Class Attributes
2 scopes for attributes

instance = variable that belongs to a specific instance or instantiation. Defined inside the constructor (init). Can access from scope of an object(local)

Class attribute = type of variable that belongs to the class itself. Defined outside the constructor. Similar to static attributes in Java (not exactly the same). Both a property of a class AND a property of the object. We can mutate a class attribute to be an instance attribute
