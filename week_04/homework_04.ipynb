```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_df = pd.read_csv('CSCI_1070_survey_results.csv')
data_df.head(22)
```

```python
data_df['What state/country were you born in?'].value_counts()
```

```python
data_df['state/country']= np.where(data_df['What state/country were you born in?']== 'Missouri',1,
                                   np.where(data_df['What state/country were you born in?']== 'United States, Missouri',2,
                                            np.where(data_df['What state/country were you born in?']=='Saint Louis, Missouri',3,
                                                     np.where(data_df['What state/country were you born in?']== 'Missouri/U.S',4,5))
                                   ))

data_df
```

```python
data_df['age'] = np.where(data_df['How old are you?']== '18', 1,
                          np.where(data_df['How old are you?']=='19',2,
                                   np.where(data_df['How old are you?']== '20',3,
                                            np.where(data_df['How old are you?']=='21',4,5))))
data_df
```

```python
data_df['What is your favorite type of food?'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
class_labels = LabelEncoder()
data_df['food'] = class_labels.fit_transform(data_df['What is your favorite type of food?'].values)
data_df['pets'] = class_labels.fit_transform(data_df['How many pets and of which kind do you have?'].values)
data_df['major'] = class_labels.fit_transform(data_df['What is your major?'].values)

data_df
```

```python
state_country = data_df.groupby('Are you working while in school?').count()
state_country
```

```python
colors = ['red','orange']
state_country.plot(kind='pie',y='Where do you live?',autopct = '%1.0f%%', colors =colors)
```
