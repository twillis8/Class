```python
import numpy as np      
import pandas as pd

belize_listing_path = "belize_airbnb_listings.csv"
belize_df = pd.read_csv(belize_listing_path)
belize_df.head()
```

```python
my_df = pd.DataFrame(belize_df)
my_df.head()
```

```python
my_df[['price','number_of_reviews','availability_365']].agg(['mean','sum','max'])
grouped_neighborhood = my_df.groupby('neighbourhood').agg({'price': 'mean','number_of_reviews':'sum','availability_365':'max'})
grouped_neighborhood
```
